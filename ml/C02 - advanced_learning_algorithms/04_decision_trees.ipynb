#### Measuring purity

The formula for the entropy function can be written as:

$H(X) = \sum_{i=1}^{n} -P(x_i) \log_2(P(x_i))$


Where:
- $H(X)$ represents the entropy of the random variable X.
- $P(x_i)$ represents the probability of the ith outcome of X.
- $n$ represents the total number of possible outcomes of X.
